```python
import kagglehub
from kagglehub import KaggleDatasetAdapter
import pandas as pd
import os
import requests
import zipfile
from io import BytesIO
```

```python
# Download latest version
path = kagglehub.dataset_download("ahmedshahriarsakib/usa-real-estate-dataset")
```

```python
for file in os.listdir(path):
    if file.endswith(".csv"):
        csv_path = os.path.join(path, file)
        print("Loading:", csv_path)
        df = pd.read_csv(csv_path)
        break
```

```python
#Save houses df to working directory
df.to_csv('Houses_Sold.csv')
```

```python
zip_url = "https://hazards.fema.gov/nri/Content/StaticDocuments/DataDownload//NRI_Table_Counties/NRI_Table_Counties.zip"
os.makedirs(".",exist_ok=True)
resp = requests.get(zip_url)
resp.raise_for_status()
with BytesIO(resp.content) as b:
    with zipfile.ZipFile(b) as z:
        z.extractall()
```

```python
path = kagglehub.dataset_download("shaswotj/us-population-by-zip-code")
```

```python
for file in os.listdir(path):
    if file.endswith(".csv"):
        csv_path = os.path.join(path, file)
        print("Loading:", csv_path)
        df = pd.read_csv(csv_path)
        break
df.columns = df.iloc[0]
df = df[1:].reset_index(drop=True)
```

```python
df.to_csv('zip_pop.csv')
```
