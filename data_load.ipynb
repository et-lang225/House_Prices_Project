```python
import requests
import pandas as pd
# (Use your actual API key or a variable)
url = ("https://api.census.gov/data/2023/acs/acs5/profile"
       "?get=group(DP05)&for=zip%20code%20tabulation%20area:*"
       "&key=2ac3fdb294a41d863f6d7e206982fe0b83bdf248")
resp = requests.get(url)
resp.raise_for_status()
data = resp.json() 
header = data[0]
records = data[1:]
Pop_Data = pd.DataFrame(records, columns=header)

# convert numeric columns from strings to ints/floats
for col in Pop_Data.columns:
    if col not in ("zip code tabulation area",):
        Pop_Data[col] = pd.to_numeric(Pop_Data[col], errors="coerce")
Pop_Data.to_csv('Zip_Pop.csv')
```
